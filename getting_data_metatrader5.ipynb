# Getting data from Metatrader5

```python
import MetaTrader5 as mt5
mt5.initialize(path = r"C:\Program Files\STP Trading MT5 Terminal\terminal64.exe")
```

```python
import pandas as pd
import datetime
current_time = datetime.datetime.now()
rates = mt5.copy_rates_from("USDJPY.stp", mt5.TIMEFRAME_D1, current_time, 100)
rates_frame = pd.DataFrame(rates)
rates_frame
```

```python
rates_frame['time'] = pd.to_datetime(rates_frame['time'], unit='s')
rates_frame.set_index('time', inplace=True)
rates_frame
```

```python
rates_frame.to_csv('USDJPY_D1.csv')
```

# Trading

```python
symbol = "USDJPY.stp"
lot = 0.01
point = mt5.symbol_info(symbol).point
price = mt5.symbol_info_tick(symbol).ask # or bid
request = {
    "action": mt5.TRADE_ACTION_DEAL,
    "symbol": symbol,
    "volume": lot,
    "type":mt5.ORDER_TYPE_BUY, # or ORDER_TYPE_SELL
    "price": price,
    "sl": price - 100 * point,
    "tp": price + 300 * point,
    "magic": 234000,
    "comment": "python script open position",
    "type_time": mt5.ORDER_TIME_GTC,
    "type_filling": mt5.ORDER_FILLING_IOC,
}
mt5.order_send(request)
```

```python
position_id = 679940
price = mt5.symbol_info_tick(symbol).bid # or ask
request = {
    "action": mt5.TRADE_ACTION_DEAL,
    "symbol": symbol,
    "volume": lot,
    "type":mt5.ORDER_TYPE_SELL, # or ORDER_TYPE_BUY
    "position": position_id,
    "price": price,
    "magic": 234000,
    "comment": "python script close position",
    "type_time": mt5.ORDER_TIME_GTC,
    "type_filling": mt5.ORDER_FILLING_IOC,
}
mt5.order_send(request)
```
